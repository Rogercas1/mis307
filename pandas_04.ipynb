### Select rows using loc and iloc

````python
housing_df.loc[0:3]
````

````python
housing_df.iloc[0:4]
````

````python
housing_df.loc[10:20]
````

````python
housing_df.iloc[10:21]
````

````python
housing_df.loc[-10:] # note that this doesn't work !
````

## Select Columns


### Renaming columns

The names of columns can often be long, not descriptive, or inconvenient to write. Columns names can also have hidden space. 


````python
housing_df.columns
````

To rename a column, you can either explicitly rename a single column, or you could write Python code to search for patterns in names and programatically rename a set of columns. 

#### Explicit renaming of a column

````python
housing_df = housing_df.rename(columns={'TOTAL VALUE ': 'TOTAL_VALUE'})# explicit
````

#### Dynamically (programatically) renaming multiple columns


````python
housing_df.columns = [s.strip().replace(' ','_') for s in housing_df.columns] # list comprehension, all columns
````

#### housing_df.iloc[4:10][:, 0:10] <- why this doesn't work

Notice that a statement of the form ```housing_df.iloc[4:10][:, 3:6]``` doesn't work. 

````python
housing_df.iloc[4:10][:, 3:6]
````

Why doesn't the above code work?

Note that there a two major parts to this statment. The first statement ```housing_df.iloc[4:10]``` produces a datafame, as folllows...

````python
housing_df.iloc[4:10]
````

Let's store this result in a temporary data frame...

````python
tmp_df = housing_df.iloc[4:10]
tmp_df
````

Now, see what happens when we try to do the following...

````python
tmp_df[:, 3:6] 
````

This is because the default processing for a dataframe slice is to use the loc style, that is, use labels (not index numbers) to refer to columns and rows. The code above is trying to use the index form (iloc) to reference columns, and therefore doesn't work in this content. 

To make the the statement ```housing_df.iloc[4:10][:, 0:10]``` work, we need to do something like...

````python
housing_df.iloc[4:10].iloc[:, 3:6]
````
